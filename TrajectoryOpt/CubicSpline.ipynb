```python
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(6)
# Points: (0,0,0), (3,3,3), (8,4,2), (5,2,7),(3,9,6),(0,0,0)
points = np.array([[0, 0, 0], [3, 3, 3], [8, 4, 2], [5, 2, 7], [3, 9, 6], [0, 0, 0]])
cs = CubicSpline(t, points)
```

```python
xs = np.linspace(0, 5, 100)

ax = plt.figure().add_subplot(projection='3d')
ax.set_xlim(-0.5, 10)
ax.set_ylim(-0.5, 10)
ax.set_zlim(-0.5, 10)
plt.plot(points[:,0], points[:,1], points[:,2], "o-")
plt.plot(cs(xs)[:,0], cs(xs)[:,1], cs(xs)[:,2])

ax.view_init(elev=20., azim=0, roll=0)
plt.show()
```

```python
plt.plot(t, points[:,0], 'o')
plt.plot(xs, cs(xs)[:,0], label="S")
plt.plot(xs, cs(xs, 1)[:,0], label="S'")
plt.plot(xs, cs(xs, 2)[:,0], label="S''")
plt.plot(xs, cs(xs, 3)[:,0], label="S'''")
plt.legend(loc='lower right', ncol=2)
plt.show()
```
